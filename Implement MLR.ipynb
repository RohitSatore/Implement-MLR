```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, metrics
```

```python
bos = datasets.load_boston(return_X_y = False)
```

```python
X = bos.data
y = bos.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.5, random_state = 0)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
print('Coefficients: \n', reg.coef_)
print('Variance score: {}'.format(reg.score(X_test, y_test)))
plt.style.use('fivethirtyeight')
plt.scatter(reg.predict(X_train), reg.predict(X_train) - y_train, color = "red", s = 5, label = 'Train data')
plt.scatter(reg.predict(X_test), reg.predict(X_test) - y_test, color = "blue", s = 5, label = 'Test data')
plt.hlines(y = 0, xmin = 0, xmax = 70, linewidth = 1)
plt.legend(loc = 'upper right')
plt.title("Residual errors")
plt.show()
```
